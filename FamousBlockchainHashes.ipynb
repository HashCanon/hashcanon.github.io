### 1. Bitcoin Genesis Block — January 3, 2009

The first block of the network (Block 0) was hardcoded into the Bitcoin Core software and forms the root of the entire chain. Its coinbase transaction famously includes the newspaper headline:
*“The Times 03/Jan/2009 Chancellor on brink of second bailout for banks”*
— a manifesto-comment by Satoshi on the collapse of the banking system.

The block minted 50 BTC, but due to a special script, the reward is unspendable. Over 25 million subsequent blocks link to this one via hash-chain ancestry.
([Bitcoin Wiki][1], [Investopedia][2])

### 2. Ethereum Genesis Block — July 30, 2015

Launched at **15:26 UTC** on 30 July 2015, Ethereum’s block 0 established an initial supply of **≈72 million ETH** — 60 million allocated to 2014 crowdsale participants and 12 million to the core-developer fund. Its hash `0xd4e5…fa3` became the sync-anchor for all full nodes and locked in the network parameters (gas limit, difficulty, etc.). From that moment, smart contracts were live on mainnet, eventually paving the way for DeFi and NFTs. ([Medium][3])

> **Note on Ethereum PoW block hashes.**  
> In Bitcoin, the Proof-of-Work condition is applied directly to the block header hash (`double-SHA256(header) ≤ target`), which often yields visible leading zeros in the published `block hash`.  
> In pre-Merge Ethereum, PoW (Ethash) validated `mixHash` and `nonce` against a target, while the recorded `block hash` is `keccak256(header)`; it is not required to show leading zeros.  
> As a result, Ethereum PoW blocks typically look “random” as plain hashes, even though their PoW still enforced difficulty via Ethash.

[3]: https://medium.com/%40mantri-nodes/exploring-ethereums-genesis-block-2d28fd282fb8 "Exploring Ethereum's Genesis Block | by Mantri Nodes | Medium"

### 3. DAO Fork – Block 1,920,000 (ETH Chain)

On July 17, 2016, the community voted to roll back the theft of 3.6 million ETH from The DAO fund. At block height 1,920,000, a hard fork was implemented: the stolen funds were redirected into a special refund contract. The hash of the main fork block `0x4985…11bb` sealed this “social” decision, and the chain that retained this change became known as **ETH**, the Ethereum we know today. ### 5. Terminal PoW Block of the Merge — No. 15,537,393 (September 15, 2022)

When the **total terminal difficulty** reached `5.875 × 10²²`, block **15 537 393** became the last one mined via Proof-of-Work (F2Pool). Its hash defined the “Paris” transition point, after which clients switched to Proof-of-Stake: every subsequent execution payload arrived from the Beacon Chain. Network-wide energy consumption fell by **≈ 99.9 %**, and mining rewards dropped to zero. ([eth2book.info][6], [SSRN][7])

### 9. The “Zero Address” — Where Tokens Are Burned

In Ethereum (and most EVM-compatible blockchains), the address  
`0x0000000000000000000000000000000000000000` exists as a valid recipient — **but it has no private key**. Sending ETH, ERC-20 tokens, or NFTs to this address means that no one — not the sender, miners, or client developers — can ever sign a transaction “back.” That’s why transfers to the zero address are considered **irreversible burns**: the assets are permanently removed from circulation, reducing the circulating supply. ([Cointelegraph][1], [Blockchain Magazine][2])

The earliest “cultural” example is linked to community practices: ERC-20 token supplies were often hard-capped by sending excess tokens to `0x0…0`, since the standard lacked a dedicated `burn()` function for a long time. The same logic applies to NFTs today: if a creator wants to destroy a token, they send it to the zero address (or to the symbolic variant `0x000…dEaD`). In this way, the blockchain’s “black hole” becomes a kind of **crypto-void** — a deterministic graveyard for forgotten projects, marketing airdrops, or intentional sacrifices. ([Ethereum Stack Exchange][3], [Crypto Council for Innovation][4])

[1]: https://cointelegraph.com/explained/what-is-the-ethereum-eth-burn-address "What is the Ethereum (ETH) burn address - Cointelegraph"
[2]: https://blockchainmagazine.net/guide-to-ethereum-burn-address "Your Ultimate Guide To Ethereum Burn Address And Top 5 Reasons ..."
[3]: https://ethereum.stackexchange.com/questions/134889/where-does-burnt-ether-go "Where does burnt ether go? - Ethereum Stack Exchange"
[4]: https://cryptoforinnovation.org/what-is-nft-burning "What is NFT Burning? - Crypto Council for Innovation"

### 10. The “Dead” Address — A Ritual Grave for Tokens

In most EVM chains, the community-defined burn address **0x000000000000000000000000000000000000dEaD** (often shortened to **0x…dEaD**) plays the same practical role as the zero address: assets sent here can never be spent again because no private key exists. What distinguishes it is its *semantic payload*: the suffix “dEaD” makes the act of destruction explicit, turning a purely technical sink into a symbolic grave. Block-explorer interfaces such as Etherscan and many analytics APIs automatically exclude balances at 0x…dEaD from “circulating supply” counts, treating them as provably burned. ([etherscan.io][1])

### 11. The “dead-42069” Vanity Burn Address

**Address:** `0xdead000000000000000042069420694206942069`

— **Functionally identical to other burn sinks.**  Like `0x…dead`, this address has no known private key; transfers here are cryptographically irreversible and therefore remove tokens (or ETH) from circulation. Explorer back-ends and indexers tag all incoming transfers as *burns* and exclude the balance from supply metrics. ([shibburn.com][1])

— **Canonicalised by the 2021 SHIB bonfire.**  On 16 May 2021, Vitalik Buterin sent \~410 trillion SHIB (≈ US \$6.7 billion) to this vanity address, instantly making it the community’s default burn target. ([coindesk.com][2])

— **Adopted well beyond SHIB.**  Meme-coin projects, NFT creators and even individual holders use the same sink for demonstrative token destruction; notable cases include the wallet *nd4.eth*, which in 2023 burned 2 500 ETH (\~US \$4.5 million) to this address. ([binance.com][3])

— **Indexed as “Burn Address 1”.**  Trackers such as Shibburn and Dune dashboards surface it separately, allowing analysts to monitor aggregate burn volumes without relying on each contract’s internal `burn()` events. ([shibburn.com][1])

In practice, **`0xdead…42069` has become Ethereum’s most recognisable public furnace** — a single, deterministic endpoint for provably sending assets out of existence while leaving an indelible audit trail.
